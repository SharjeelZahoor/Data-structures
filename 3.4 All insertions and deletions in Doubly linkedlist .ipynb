Inserting Value at start in Doubly linkedlist and print list after adding values

```python
class Node:
  def __init__(self, data):
    self.data = data
    self.prev = None
    self.next = None

class DoublyLinkedList:

  def __init__(self):
    self.head = None

#insertion at end and deletion from start
  def insertAtStart(self, data):
    new_node = Node(data)
    if self.head is None:
      self.head = new_node
    else:
      new_node.next = self.head
      self.head.prev = new_node
      self.head = new_node

  def deleteFromStart(self):
    if self.head is None:
      return "List is empty"
    elif self.head.next is None:
      print("deleted node : ", self.head.data)
      self.head = None
    else:
      print("deleted node : ", self.head.data)
      self.head.next.prev = None
      self.head = self.head.next

# insertion at end and deletion from end
  def insertAtEnd(self, data):
    new_node = Node(data)
    if self.head is None:
      self.head = new_node
    else:
      temp = self.head
      while temp.next:
        temp = temp.next
      temp.next = new_node
      new_node.prev = temp
  def deleteFromEnd(self):
    if self.head is None:
      return "List is empty"
    elif self.head.next is None:
      print("deleted node : ", self.head.data)
      self.head = None
    else:
      temp = self.head
      while temp.next.next:
        temp = temp.next
      print("deleted node : ", temp.next.data)
      temp.next = None

#insertion and deletion at nth position
  def insertAtNthIndex(self, data, n):
    new_node = Node(data)
    if self.head is None:
      self.head = new_node
    elif n == 0:
      new_node.next = self.head
      self.head.prev = new_node
      self.head = new_node
    else:
      temp = self.head
      count = 0
      while temp and count < n - 1:
        temp = temp.next
        count += 1
      if temp is None:
        return "Index not available"
      elif temp.next is None:
        temp.next = new_node
        new_node.prev = temp
      else:
        new_node.next = temp.next
        new_node.prev = temp
        temp.next.prev = new_node
        temp.next = new_node

  def deleteFromNthIndex(self, n):
    if self.head is None:
      return "List is empty"
    elif n == 0:
      self.head.next.prev = None
      self.head = self.head.next
    else:
      temp = self.head
      count = 0
      while temp and count < n - 1:
        temp = temp.next
        count += 1
      if temp is None:
        return "Index not available"
      elif temp.next is None:
        return "Index not availale"
      else:
        print("deleted node : ", temp.next.data)
        temp.next = temp.next.next
        if temp.next:
          temp.next.prev = temp


#printing list
  def printList(self):
    list2=[]
    temp = self.head
    while temp:
      list2.append(temp.data)
      temp = temp.next
    print(list2)

l1 = DoublyLinkedList()
l1.insertAtStart(10)
l1.insertAtStart(20)
l1.insertAtStart(30)
l1.printList()
l1.deleteFromStart()
l1.printList()
l1.insertAtEnd(40)
l1.insertAtEnd(50)
l1.printList()
l1.deleteFromEnd()
l1.printList()
l1.insertAtNthIndex(60, 0)
l1.insertAtNthIndex(70, 4)
l1.printList()
l1.deleteFromNthIndex(2)
l1.printList()

```
